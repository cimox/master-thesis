```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Read file with prequential accuracy [synthetic]

```python
hyp1 = pd.read_csv('hyp1.csv')
hyp1 = hyp1[hyp1['accuracy'] > 0].divide(100)

hyp10 = pd.read_csv('hyp10.csv')
hyp10 = hyp10[hyp10['accuracy'] > 0].divide(100)

hyp100 = pd.read_csv('hyp100.csv')
hyp100 = hyp100[hyp100['accuracy'] > 0].divide(100)

rbf = pd.read_csv('rbf.csv')
rbf = rbf[rbf['accuracy'] > 0].divide(100)

sea = pd.read_csv('sea.csv')
sea = sea[sea['accuracy'] > 0].divide(100)
```

### Plot whole prequential accuracy progress [synthetic]

```python
%matplotlib inline

hyp1_ax = hyp1['accuracy'].plot()
hyp10_ax = hyp10['accuracy'].plot()
hyp100_ax = hyp100['accuracy'].plot()
rbf_ax = rbf['accuracy'].plot()
sea_ax = sea['accuracy'].plot(kind='line')
sea_ax.set_xlabel("Vzorka")
sea_ax.set_ylabel("Presnosť")


plt.legend(['Hyp_1', 'Hyp_10', 'Hyp_100', 'SEA_ND', 'RBF'], loc='lower right')
plt.savefig('500k.pdf', bbox_inches='tight')
```

### Plot first 5k samples [synthetic]

```python
%matplotlib inline
k = 7053

hyp1_ax = hyp1['accuracy'].head(k).plot()
hyp10_ax = hyp10['accuracy'].head(k).plot()
hyp100_ax = hyp100['accuracy'].head(k).plot()
rbf_ax = rbf['accuracy'].head(k).plot()
sea_ax = sea['accuracy'].head(k).plot(kind='line')
sea_ax.set_xlabel("Vzorka")
sea_ax.set_ylabel("Presnosť")


plt.legend(['Hyp_1', 'Hyp_10', 'Hyp_100', 'SEA_ND', 'RBF'], loc='lower right')
plt.savefig('{}k.pdf'.format(k), bbox_inches='tight')
```

### Read file with prequential accuracy [real datasets]

```python
telco = pd.read_csv('telco.csv')
telco = telco[telco['accuracy'] > 0].divide(100)

air = pd.read_csv('air.csv')
air = air[air['accuracy'] > 0].divide(100)
```

### Plot whole prequential accuracy progress [real datasets]

```python
%matplotlib inline

telco_ax = telco['accuracy'].plot()
air_ax = air['accuracy'].plot(kind='line')

sea_ax.set_xlabel("Vzorka")
sea_ax.set_ylabel("Presnosť")


plt.legend(['Telco', 'Airlines'], loc='lower right')
plt.savefig('real_all.pdf', bbox_inches='tight')
```
